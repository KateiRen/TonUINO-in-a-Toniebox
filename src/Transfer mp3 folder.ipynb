# Transfer mp3 folder 

Transfer a folder of mp3 files (e.g. a folder with all songs of an album) to a new TonUINO folder.

- Creates the next two digit numbered folder.
- Sorts all mp3-files found by name
- Copies and renames mp3 files to three digit numbered files.
 

```python
from ipywidgets import Button
from tkinter import Tk, filedialog
from IPython.display import clear_output, display
import os
import shutil

sourcePath = "" # hard code your path if needed
destinationPath = "" # hard code your path if needed

def select_sourceDir(b):
    global sourcePath
    clear_output()
    root = Tk()
    root.withdraw() # Hide the main window.
    root.call('wm', 'attributes', '.', '-topmost', True) # Raise the root to the top of all windows.
    b.files = filedialog.askdirectory() 
    sourcePath = b.files

# maybe refactor to only use one callback function 
def select_destDir(b):
    global destinationPath
    clear_output()
    root = Tk()
    root.withdraw() # Hide the main window.
    root.call('wm', 'attributes', '.', '-topmost', True) # Raise the root to the top of all windows.
    b.files = filedialog.askdirectory() 
    destinationPath = b.files


def isValidMP3Folder(name):
    #name = name[:2] # DFPlayer
    if len(str(name))>2:
        print("Folder name should be a two digit number")
        return False
    try:
        number = int(name)
        return True
    except:
        print("Folder name could not be matched to a two digit number")
        return False

def isValidMP3File(name):
    if name[-4:] != ".mp3":
        return False
    name = name[:3] # if naming with appended song title is used
    try:
        number = int(name)
        return True
    except:
        return False
```

```python
# Select source and target paths

sourceselect = Button(description="Select Source")
sourceselect.on_click(select_sourceDir)

destselect = Button(description="Select TonUINO")
destselect.on_click(select_destDir)

display(sourceselect)
display(destselect)
```

```python
# find biggest numbered mp3 folder name
    
number  = 0
for entry in os.scandir(destinationPath):
#     print(entry.name)
    if entry.is_dir() and isValidMP3Folder(entry.name):
        if int(entry.name) > number:
            number = int(entry.name)

if number == 0:
    print("No msic folder (two digit number) found.")    
else:
    print("Highest music folder found: {0:02d}".format(number))

number += 1


# create new folder

destinationDirectory = destinationPath+"/"+"{0:02d}".format(number)
os.mkdir(destinationDirectory)
print("Created new folder: " + destinationDirectory)
```

```python
# collect mp3 files and sort

def getName(item):
    return item.name.upper() # alles in Gro√übuchtstaben wandeln

mp3files = []

for mp3 in os.scandir(sourcePath):
    if mp3.name[-4:] == ".mp3":
        mp3files.append(mp3)        

mp3files.sort(key = getName)


# copy mp3 files and rename as three digit number

number = 1 # start with 001.mp3
for mp3 in mp3files:
    shutil.copy(sourcePath + "/" + mp3.name, destinationDirectory + "/{0:03d}.mp3".format(number))
    print("Copied " + sourcePath + "/" + mp3.name)
    print("To " + destinationDirectory + "/{0:03d}.mp3".format(number)+"\n")
    number += 1
```
